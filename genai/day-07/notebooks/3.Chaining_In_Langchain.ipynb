##### Prompt Templates-Dynamic Prompts

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
```

```python
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
prompts=ChatPromptTemplate.from_messages([
    {"role":"system","content":"You are a {language} developer"},
    {"role":"user","content":"{query}"}
])
```

```python
final_prompt=prompts.format_messages(language="javascript",query="mapping and filter difference")
```

```python
res=llm.invoke(final_prompt)
print(res.content)
```

##### Chaining in Langchain

```python
from langchain_google_genai import  ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash")
output=StrOutputParser()
```

```python
prompt=ChatPromptTemplate.from_messages([
    {"role":"system","content":"You are a senior {role} developer with 6 years experience"},
    {"role":'user',"content":"{query}"}
])
```

```python
chaining=prompt | llm | output
res=chaining.invoke({"role":"game","query":"How develop a game play only when online"})
print(res)
```
