##### Connect With Google Gemini API

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
res=llm.invoke("what is fastapi")
print(res.content)
```

##### Connect With Groq API

```python
from langchain_groq import ChatGroq
llm=ChatGroq(model="llama-3.3-70b-versatile")
```

```python
res=llm.invoke("who is god of cricket")
print(res.content)
```

##### Static Prompts

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
prompts=[
    {"role":"system","content":"You are a senior python-backend developer"},
    {"role":"user","content":"howwe can apply authentication in fastapi?"}
]
```

```python
res=llm.invoke(prompts)
print(res.content)
```
