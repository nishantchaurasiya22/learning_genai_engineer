```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
prompts=[
    {"role":"system","content":"You are a translator and you are translate in hindi not explain only convert english sentence in hindi "},
    {"role":"user","content":"what is array in javascript"}
]

res=llm.invoke(prompts)
print(res.content)
```

##### Prompt Templates-Dynamic Prompts

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
```

```python
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
prompts=ChatPromptTemplate.from_messages([
    {"role":"system","content":"You are a translator, translate input in {language}"},
    {"role":"user","content":"{query}"}
])
```

```python
final_prompt=prompts.format_messages(language="hinglish",query="I Love python and javascript")
```

```python
res=llm.invoke(final_prompt)
print(res.content)
```
